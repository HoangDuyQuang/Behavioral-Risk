```python
def build_obs_features(
    transactions_df: pd.DataFrame,
    cards_df: pd.DataFrame,
    users_df: pd.DataFrame,
    snapshot_date: pd.Timestamp,
    obs_days: int = 180
) -> pd.DataFrame:
    
    """
    Build observation-window features at client_id level.
    Observation window: [snapshot_date - obs_days, snapshot_date]
    """

    obs_start = snapshot_date - pd.Timedelta(days=obs_days)

    obs_df = transactions_df.loc[
        (transactions_df['date'] > obs_start) &
        (transactions_df['date'] <= snapshot_date)
    ].copy()

    if obs_df.empty:
        return pd.DataFrame(columns=['client_id'])

    # -------------------------
    # Basic transaction flags
    # -------------------------
    obs_df['tr_overnight'] = obs_df['date'].dt.hour.between(0, 5).astype(int)
    obs_df['is_weekend'] = obs_df['date'].dt.dayofweek.isin([5, 6]).astype(int)

    obs_df['insufficient_funds'] = obs_df['errors'].fillna('No Error').str.contains(
        'Insufficient Balance', case=False, na=False
    ).astype(int)

    obs_df['has_error'] = (
        obs_df['errors'].fillna('No Error').ne('No Error')
    ).astype(int)

    high_risk_mccs = [7995, 4829]
    obs_df['high_risk_mcc'] = obs_df['mcc'].isin(high_risk_mccs).astype(int)

    obs_df['is_chip'] = obs_df['use_chip'].fillna('').str.contains('Chip', case=False, na=False).astype(int)
    obs_df['is_swipe'] = obs_df['use_chip'].fillna('').str.contains('Swipe', case=False, na=False).astype(int)
    obs_df['is_online'] = obs_df['use_chip'].fillna('').str.contains('Online', case=False, na=False).astype(int)

    obs_df['tx_month'] = obs_df['date'].dt.to_period('M')
    obs_df['tx_day'] = obs_df['date'].dt.date

    # -------------------------
    # Core aggregates
    # -------------------------
    obs_features = (
        obs_df.groupby('client_id')
        .agg(
            txn_count=('amount', 'count'),
            total_spent=('amount', 'sum'),
            avg_txn=('amount', 'mean'),
            median_txn=('amount', 'median'),
            max_txn=('amount', 'max'),
            min_txn=('amount', 'min'),
            txn_std=('amount', 'std'),

            merchant_diversity=('mcc', 'nunique'),
            merchant_count=('merchant_id', 'nunique'),
            city_diversity=('merchant_city', 'nunique'),
            state_diversity=('merchant_state', 'nunique'),

            active_days=('tx_day', 'nunique'),
            min_date=('date', 'min'),
            max_date=('date', 'max'),

            insufficient_funds_count=('insufficient_funds', 'sum'),
            error_count=('has_error', 'sum'),
            overnight_count=('tr_overnight', 'sum'),
            weekend_count=('is_weekend', 'sum'),
            high_risk_count=('high_risk_mcc', 'sum'),

            chip_count=('is_chip', 'sum'),
            swipe_count=('is_swipe', 'sum'),
            online_count=('is_online', 'sum')
        )
        .reset_index()
    )

    obs_features['txn_std'] = obs_features['txn_std'].fillna(0)

    obs_features['observed_span_days'] = (
        (obs_features['max_date'] - obs_features['min_date']).dt.days + 1
    ).clip(lower=1)

    # -------------------------
    # Ratio features
    # -------------------------
    obs_features['txn_frequency'] = obs_features['txn_count'] / obs_features['observed_span_days']
    obs_features['spending_velocity'] = obs_features['total_spent'] / obs_features['observed_span_days']

    obs_features['spending_volatility'] = (
        obs_features['txn_std'] / obs_features['avg_txn'].replace(0, np.nan)
    ).replace([np.inf, -np.inf], 0).fillna(0)

    obs_features['max_to_avg_ratio'] = (
        obs_features['max_txn'] / obs_features['avg_txn'].replace(0, np.nan)
    ).replace([np.inf, -np.inf], 0).fillna(0)

    obs_features['insufficient_funds_ratio'] = (
        obs_features['insufficient_funds_count'] / obs_features['txn_count']
    ).fillna(0)

    obs_features['declined_ratio'] = (
        obs_features['error_count'] / obs_features['txn_count']
    ).fillna(0)

    obs_features['night_ratio'] = (
        obs_features['overnight_count'] / obs_features['txn_count']
    ).fillna(0)

    obs_features['weekend_ratio'] = (
        obs_features['weekend_count'] / obs_features['txn_count']
    ).fillna(0)

    obs_features['high_risk_ratio'] = (
        obs_features['high_risk_count'] / obs_features['txn_count']
    ).fillna(0)

    obs_features['chip_ratio'] = (
        obs_features['chip_count'] / obs_features['txn_count']
    ).fillna(0)

    obs_features['swipe_ratio'] = (
        obs_features['swipe_count'] / obs_features['txn_count']
    ).fillna(0)

    obs_features['online_ratio'] = (
        obs_features['online_count'] / obs_features['txn_count']
    ).fillna(0)

    obs_features['merchant_per_txn_ratio'] = (
        obs_features['merchant_count'] / obs_features['txn_count']
    ).fillna(0)

    # -------------------------
    # Merchant entropy
    # -------------------------
    merchant_entropy_feature = (
        obs_df.groupby('client_id')['mcc']
        .apply(lambda x: entropy(x.value_counts(normalize=True)))
        .rename('merchant_entropy')
        .reset_index()
    )

    obs_features = obs_features.merge(
        merchant_entropy_feature,
        on='client_id',
        how='left'
    )
    obs_features['merchant_entropy'] = obs_features['merchant_entropy'].fillna(0)

    # -------------------------
    # Monthly behavior
    # -------------------------
    monthly_client = (
        obs_df.groupby(['client_id', 'tx_month'])
        .agg(
            monthly_spent=('amount', 'sum'),
            monthly_txn=('amount', 'count')
        )
        .reset_index()
    )

    monthly_stats = (
        monthly_client.groupby('client_id')
        .agg(
            avg_monthly_spent=('monthly_spent', 'mean'),
            std_monthly_spent=('monthly_spent', 'std'),
            avg_monthly_txn=('monthly_txn', 'mean'),
            std_monthly_txn=('monthly_txn', 'std'),
            active_months=('tx_month', 'nunique')
        )
        .reset_index()
    )

    monthly_stats['std_monthly_spent'] = monthly_stats['std_monthly_spent'].fillna(0)
    monthly_stats['std_monthly_txn'] = monthly_stats['std_monthly_txn'].fillna(0)

    def calc_spend_trend(x):
        x = x.sort_values('tx_month')
        if len(x) < 2:
            return 0
        y = x['monthly_spent'].values
        t = np.arange(len(y))
        return np.polyfit(t, y, 1)[0]

    monthly_trend = (
        monthly_client.groupby('client_id')
        .apply(calc_spend_trend)
        .rename('monthly_spend_trend')
        .reset_index()
    )

    obs_features = obs_features.merge(monthly_stats, on='client_id', how='left')
    obs_features = obs_features.merge(monthly_trend, on='client_id', how='left')

    # -------------------------
    # Transaction gap
    # -------------------------
    gap_df = obs_df.sort_values(['client_id', 'date']).copy()
    gap_df['txn_gap'] = gap_df.groupby('client_id')['date'].diff().dt.days

    gap_features = (
        gap_df.groupby('client_id')
        .agg(
            avg_txn_gap_days=('txn_gap', 'mean'),
            std_txn_gap_days=('txn_gap', 'std')
        )
        .reset_index()
    )

    obs_features = obs_features.merge(gap_features, on='client_id', how='left')

    # -------------------------
    # Recency
    # -------------------------
    last_txn = (
        obs_df.groupby('client_id')['date']
        .max()
        .rename('last_txn_date')
        .reset_index()
    )

    last_txn['recency_days'] = (snapshot_date - last_txn['last_txn_date']).dt.days

    obs_features = obs_features.merge(
        last_txn[['client_id', 'recency_days']],
        on='client_id',
        how='left'
    )

    # -------------------------
    # Card features
    # -------------------------
    card_features = (
        cards_df.groupby('client_id')
        .agg(
            total_credit_limit=('credit_limit', 'sum'),
            avg_credit_limit=('credit_limit', 'mean'),
            card_count=('id', 'count'),
            debit_card_count=('card_type', lambda x: x.fillna('').str.contains('Debit', case=False).sum()),
            credit_card_count=('card_type', lambda x: x.fillna('').str.contains('Credit', case=False).sum()),
            chip_card_count=('has_chip', lambda x: (x == 'YES').sum()),
            avg_cards_issued=('num_cards_issued', 'mean')
        )
        .reset_index()
    )

    obs_features = obs_features.merge(card_features, on='client_id', how='left')

    # -------------------------
    # User features
    # -------------------------
    user_features = users_df[[
        'id', 'current_age', 'retirement_age', 'per_capita_income',
        'yearly_income', 'total_debt', 'credit_score', 'num_credit_cards',
        'debt_to_income', 'age_to_retirement'
    ]].rename(columns={'id': 'client_id'})

    obs_features = obs_features.merge(user_features, on='client_id', how='left')

    # -------------------------
    # Cross features
    # -------------------------
    obs_features['spend_to_limit_ratio'] = (
        obs_features['total_spent'] / obs_features['total_credit_limit'].replace(0, np.nan)
    ).replace([np.inf, -np.inf], 0).fillna(0)

    obs_features['spend_to_income_ratio'] = (
        obs_features['total_spent'] / obs_features['yearly_income'].replace(0, np.nan)
    ).replace([np.inf, -np.inf], 0).fillna(0)

    obs_features['avg_txn_to_income_ratio'] = (
        obs_features['avg_txn'] / obs_features['yearly_income'].replace(0, np.nan)
    ).replace([np.inf, -np.inf], 0).fillna(0)

    obs_features['txn_per_card'] = (
        obs_features['txn_count'] / obs_features['card_count'].replace(0, np.nan)
    ).replace([np.inf, -np.inf], 0).fillna(0)

    obs_features['spent_per_card'] = (
        obs_features['total_spent'] / obs_features['card_count'].replace(0, np.nan)
    ).replace([np.inf, -np.inf], 0).fillna(0)

    # Recent window ratios
    last_60_df = obs_df.loc[obs_df['date'] > snapshot_date - pd.Timedelta(days=60)]
    spend_60 = last_60_df.groupby('client_id')['amount'].sum().rename('spend_last_60')
    txn_60 = last_60_df.groupby('client_id')['amount'].count().rename('txn_last_60')

    last_120_df = obs_df.loc[obs_df['date'] > snapshot_date - pd.Timedelta(days=120)]
    spend_120 = last_120_df.groupby('client_id')['amount'].sum().rename('spend_last_120')
    txn_120 = last_120_df.groupby('client_id')['amount'].count().rename('txn_last_120')

    obs_features = obs_features.merge(spend_60, on='client_id', how='left')
    obs_features = obs_features.merge(txn_60, on='client_id', how='left')
    obs_features = obs_features.merge(spend_120, on='client_id', how='left')
    obs_features = obs_features.merge(txn_120, on='client_id', how='left')

    obs_features['spending_acceleration_60_120'] = (
        obs_features['spend_last_60'] / obs_features['spend_last_120'].replace(0, np.nan)
    ).replace([np.inf, -np.inf], 0).fillna(0)

    obs_features['txn_acceleration_60_120'] = (
        obs_features['txn_last_60'] / obs_features['txn_last_120'].replace(0, np.nan)
    ).replace([np.inf, -np.inf], 0).fillna(0)

    # Metadata
    obs_features['snapshot_date'] = pd.Timestamp(snapshot_date)
    obs_features['obs_start_date'] = pd.Timestamp(obs_start)
    obs_features['obs_end_date'] = pd.Timestamp(snapshot_date)

    # Cleanup
    obs_features = obs_features.drop(columns=['min_date', 'max_date'], errors='ignore')
    obs_features = obs_features.fillna(0)

    return obs_features
```

```python
ABSTAIN = -1


def _safe_div(a, b):
    return np.divide(a, b, out=np.zeros_like(a, dtype=float), where=(b != 0))


def build_proxy_base_for_snapshot(
    transactions_df: pd.DataFrame,
    snapshot_date: pd.Timestamp,
    perf_days: int = 60,
    obs_days: int = 180,
    alpha: float = 10.0
) -> pd.DataFrame:
    """
    Tạo bảng proxy future-side + context past-side ở cấp client_id.
    Đây chưa phải label, chỉ là base để các labeling functions bỏ phiếu.
    """

    perf_end = snapshot_date + pd.Timedelta(days=perf_days)
    obs_start = snapshot_date - pd.Timedelta(days=obs_days)

    perf_df = transactions_df.loc[
        (transactions_df['date'] > snapshot_date) &
        (transactions_df['date'] <= perf_end)
    ].copy()

    obs_df = transactions_df.loc[
        (transactions_df['date'] > obs_start) &
        (transactions_df['date'] <= snapshot_date)
    ].copy()

    if perf_df.empty:
        return pd.DataFrame()

    # -------------------------
    # Common flags
    # -------------------------
    for df in [perf_df, obs_df]:
        df['insufficient_funds'] = df['errors'].fillna('No Error').str.contains(
            'Insufficient Balance', case=False, na=False
        ).astype(int)

        df['has_error'] = (
            df['errors'].fillna('No Error').ne('No Error')
        ).astype(int)

        high_risk_mccs = [7995, 4829]
        df['high_risk_mcc'] = df['mcc'].isin(high_risk_mccs).astype(int)

        df['tx_day'] = df['date'].dt.date

    # -------------------------
    # Future window stats
    # -------------------------
    future_stats = (
        perf_df.groupby('client_id')
        .agg(
            future_txn_count=('amount', 'count'),
            future_spend=('amount', 'sum'),
            future_avg_txn=('amount', 'mean'),
            future_max_txn=('amount', 'max'),
            future_active_days=('tx_day', 'nunique'),

            future_insufficient_funds_count=('insufficient_funds', 'sum'),
            future_error_count=('has_error', 'sum'),
            future_high_risk_count=('high_risk_mcc', 'sum')
        )
        .reset_index()
        .set_index('client_id')
    )

    # -------------------------
    # Past window stats
    # -------------------------
    if obs_df.empty:
        past_stats = pd.DataFrame(index=future_stats.index)
    else:
        past_stats = (
            obs_df.groupby('client_id')
            .agg(
                past_txn_count=('amount', 'count'),
                past_spend=('amount', 'sum'),
                past_avg_txn=('amount', 'mean'),
                past_max_txn=('amount', 'max'),
                past_active_days=('tx_day', 'nunique'),

                past_insufficient_funds_count=('insufficient_funds', 'sum'),
                past_error_count=('has_error', 'sum'),
                past_high_risk_count=('high_risk_mcc', 'sum')
            )
            .reset_index()
            .set_index('client_id')
        )

    label_base = future_stats.join(past_stats, how='left').fillna(0)

    # -------------------------
    # Smoothed / ratio features
    # -------------------------
    global_high_risk_rate = perf_df['high_risk_mcc'].mean() if len(perf_df) > 0 else 0.0

    label_base['future_high_risk_smoothed'] = (
        label_base['future_high_risk_count'] + alpha * global_high_risk_rate
    ) / (label_base['future_txn_count'] + alpha)

    label_base['future_error_rate'] = _safe_div(
        label_base['future_error_count'].values,
        label_base['future_txn_count'].values
    )
    label_base['past_error_rate'] = _safe_div(
        label_base['past_error_count'].values,
        label_base['past_txn_count'].values
    )

    label_base['future_insufficient_rate'] = _safe_div(
        label_base['future_insufficient_funds_count'].values,
        label_base['future_txn_count'].values
    )
    label_base['past_insufficient_rate'] = _safe_div(
        label_base['past_insufficient_funds_count'].values,
        label_base['past_txn_count'].values
    )

    label_base['velocity_future'] = _safe_div(
        label_base['future_spend'].values,
        label_base['past_spend'].replace(0, np.nan).fillna(0).values
    )

    label_base['txn_velocity_future'] = _safe_div(
        label_base['future_txn_count'].values,
        label_base['past_txn_count'].replace(0, np.nan).fillna(0).values
    )

    label_base['avg_txn_shift'] = _safe_div(
        label_base['future_avg_txn'].values,
        label_base['past_avg_txn'].replace(0, np.nan).fillna(0).values
    )

    label_base['error_rate_delta'] = (
        label_base['future_error_rate'] - label_base['past_error_rate']
    )

    label_base['insufficient_rate_delta'] = (
        label_base['future_insufficient_rate'] - label_base['past_insufficient_rate']
    )

    label_base['high_risk_delta'] = (
        label_base['future_high_risk_smoothed']
        - _safe_div(
            label_base['past_high_risk_count'].values,
            np.maximum(label_base['past_txn_count'].values, 1)
        )
    )

    label_base['snapshot_date'] = pd.Timestamp(snapshot_date)
    label_base['perf_end_date'] = pd.Timestamp(perf_end)

    return label_base.reset_index()
```

```python
def build_label_matrix(label_base: pd.DataFrame) -> pd.DataFrame:
    """
    Mỗi LF trả về:
    - 1  : positive / distress
    - 0  : negative / non-distress
    - -1 : abstain
    """

    df = label_base.copy()

    q90_velocity = df['velocity_future'].quantile(0.90) if len(df) > 0 else 0
    q95_velocity = df['velocity_future'].quantile(0.95) if len(df) > 0 else 0
    q90_highrisk = df['future_high_risk_smoothed'].quantile(0.90) if len(df) > 0 else 0
    q75_highrisk = df['future_high_risk_smoothed'].quantile(0.75) if len(df) > 0 else 0
    q90_error_delta = df['error_rate_delta'].quantile(0.90) if len(df) > 0 else 0

    L = pd.DataFrame(index=df.index)

    # -------------------------
    # Positive LFs
    # -------------------------
    # LF1: strong signal - repeated insufficient funds
    L['lf_pos_repeat_if'] = np.where(
        df['future_insufficient_funds_count'] >= 2, 1, ABSTAIN
    )

    # LF2: insufficient funds + high error rate
    L['lf_pos_if_plus_error'] = np.where(
        (df['future_insufficient_funds_count'] >= 1) &
        (df['future_error_rate'] >= 0.20),
        1, ABSTAIN
    )

    # LF3: insufficient funds + spending spike
    L['lf_pos_if_plus_velocity'] = np.where(
        (df['future_insufficient_funds_count'] >= 1) &
        (df['velocity_future'] >= max(1.5, q90_velocity)),
        1, ABSTAIN
    )

    # LF4: insufficient funds + risky MCC concentration
    L['lf_pos_if_plus_risky_mcc'] = np.where(
        (df['future_insufficient_funds_count'] >= 1) &
        (df['future_high_risk_smoothed'] >= max(0.15, q90_highrisk)),
        1, ABSTAIN
    )

    # LF5: severe behavior shift without repeated IF but with many alerts
    L['lf_pos_behavioral_shock'] = np.where(
        (df['future_insufficient_funds_count'] == 0) &
        (df['velocity_future'] >= max(2.0, q95_velocity)) &
        (df['error_rate_delta'] >= max(0.10, q90_error_delta)) &
        (df['future_error_count'] >= 2),
        1, ABSTAIN
    )

    # -------------------------
    # Negative LFs
    # -------------------------
    # LF6: completely clean and stable
    L['lf_neg_clean_stable'] = np.where(
        (df['future_insufficient_funds_count'] == 0) &
        (df['future_error_count'] == 0) &
        (df['velocity_future'] >= 0.7) &
        (df['velocity_future'] <= 1.3),
        0, ABSTAIN
    )

    # LF7: low risk behavior, no IF, low risky MCC
    L['lf_neg_lowrisk_pattern'] = np.where(
        (df['future_insufficient_funds_count'] == 0) &
        (df['future_error_rate'] <= 0.05) &
        (df['future_high_risk_smoothed'] <= q75_highrisk),
        0, ABSTAIN
    )

    # LF8: inactive / low activity and clean
    L['lf_neg_low_activity_clean'] = np.where(
        (df['future_insufficient_funds_count'] == 0) &
        (df['future_error_count'] == 0) &
        (df['future_txn_count'] <= 2),
        0, ABSTAIN
    )

    return L
```

```python
def estimate_lf_weights(label_matrix: pd.DataFrame) -> pd.Series:
    """
    Ước lượng độ tin cậy tương đối của từng LF mà không cần label thật.
    Dựa trên:
    - coverage
    - agreement với phần còn lại
    """

    weights = {}

    for col in label_matrix.columns:
        lf = label_matrix[col]
        covered = lf != ABSTAIN
        coverage = covered.mean()

        if coverage == 0:
            weights[col] = 0.0
            continue

        others = label_matrix.drop(columns=[col])

        agreements = []
        for i in label_matrix.index[covered]:
            other_votes = others.loc[i]
            other_votes = other_votes[other_votes != ABSTAIN]

            if len(other_votes) == 0:
                continue

            pos_votes = (other_votes == 1).sum()
            neg_votes = (other_votes == 0).sum()

            if pos_votes == neg_votes:
                continue

            consensus = 1 if pos_votes > neg_votes else 0
            agreements.append(int(lf.loc[i] == consensus))

        agreement = np.mean(agreements) if len(agreements) > 0 else 0.5

        # weight vừa phải, tránh LF nào lấn át hoàn toàn
        weight = (0.5 + agreement) * np.sqrt(coverage)
        weights[col] = float(np.clip(weight, 0.15, 1.50))

    return pd.Series(weights).sort_values(ascending=False)


def aggregate_weak_labels(
    label_base: pd.DataFrame,
    label_matrix: pd.DataFrame,
    pos_threshold: float = 0.80,
    neg_threshold: float = 0.20
):
    """
    Tạo:
    - label_prob
    - label_confidence
    - distress_label (NaN nếu chưa đủ chắc)
    """

    lf_weights = estimate_lf_weights(label_matrix)

    weighted_pos = pd.Series(0.0, index=label_matrix.index)
    weighted_neg = pd.Series(0.0, index=label_matrix.index)
    covered_weight = pd.Series(0.0, index=label_matrix.index)

    for col in label_matrix.columns:
        w = lf_weights[col]
        vals = label_matrix[col]

        weighted_pos += np.where(vals == 1, w, 0.0)
        weighted_neg += np.where(vals == 0, w, 0.0)
        covered_weight += np.where(vals != ABSTAIN, w, 0.0)

    raw_score = np.where(
        covered_weight > 0,
        (weighted_pos - weighted_neg) / covered_weight,
        0.0
    )

    # map score [-1, 1] -> prob [0, 1]
    label_prob = 1 / (1 + np.exp(-3.0 * raw_score))
    label_confidence = np.maximum(label_prob, 1 - label_prob)

    distress_label = np.where(
        label_prob >= pos_threshold, 1,
        np.where(label_prob <= neg_threshold, 0, np.nan)
    )

    out = label_base.copy()
    out['label_prob'] = label_prob
    out['label_confidence'] = label_confidence
    out['lf_votes_pos'] = (label_matrix == 1).sum(axis=1)
    out['lf_votes_neg'] = (label_matrix == 0).sum(axis=1)
    out['lf_covered'] = (label_matrix != ABSTAIN).sum(axis=1)
    out['distress_label'] = distress_label

    return out, lf_weights
```

```python
def build_label_for_snapshot_weak(
    transactions_df: pd.DataFrame,
    snapshot_date: pd.Timestamp,
    perf_days: int = 60,
    obs_days: int = 180,
    alpha: float = 10.0,
    pos_threshold: float = 0.80,
    neg_threshold: float = 0.20
):
    label_base = build_proxy_base_for_snapshot(
        transactions_df=transactions_df,
        snapshot_date=snapshot_date,
        perf_days=perf_days,
        obs_days=obs_days,
        alpha=alpha
    )

    if label_base.empty:
        return pd.DataFrame(columns=['client_id', 'distress_label']), pd.DataFrame(), pd.Series(dtype=float)

    label_matrix = build_label_matrix(label_base)
    label_debug, lf_weights = aggregate_weak_labels(
        label_base=label_base,
        label_matrix=label_matrix,
        pos_threshold=pos_threshold,
        neg_threshold=neg_threshold
    )

    label = label_debug[['client_id', 'distress_label']].copy()

    return label, label_debug, lf_weights
```

```python
def build_snapshot_dataset(
    transactions_df: pd.DataFrame,
    cards_df: pd.DataFrame,
    users_df: pd.DataFrame,
    snapshot_date: pd.Timestamp,
    obs_days: int = 180,
    perf_days: int = 60,
    alpha: float = 10.0,
    pos_threshold: float = 0.80,
    neg_threshold: float = 0.20
):
    obs_features = build_obs_features(
        transactions_df=transactions_df,
        cards_df=cards_df,
        users_df=users_df,
        snapshot_date=snapshot_date,
        obs_days=obs_days
    )

    label, label_debug, lf_weights = build_label_for_snapshot_weak(
        transactions_df=transactions_df,
        snapshot_date=snapshot_date,
        perf_days=perf_days,
        obs_days=obs_days,
        alpha=alpha,
        pos_threshold=pos_threshold,
        neg_threshold=neg_threshold
    )

    if obs_features.empty:
        return pd.DataFrame(), pd.DataFrame(), pd.Series(dtype=float)

    dataset = obs_features.merge(label, on='client_id', how='left')

    # KHÔNG fillna(0) nữa, vì NaN nghĩa là pseudo-label chưa chắc
    dataset = dataset.merge(
        label_debug[
            [
                'client_id',
                'label_prob',
                'label_confidence',
                'lf_votes_pos',
                'lf_votes_neg',
                'lf_covered',
                'future_insufficient_funds_count',
                'future_error_count',
                'future_error_rate',
                'future_high_risk_smoothed',
                'velocity_future'
            ]
        ],
        on='client_id',
        how='left'
    )

    dataset['snapshot_date'] = pd.Timestamp(snapshot_date)

    return dataset, label_debug, lf_weights
```
